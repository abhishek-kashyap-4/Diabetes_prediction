```
import sklearn
import pandas as pd

dataset=pd.read_csv('diabetes.csv')
X=dataset.iloc[:,:-1].values
y=dataset.iloc[:,-1].values

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(X,y,test_size=0.3)

#feature scaling
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)


import keras
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout

def build_regressor():
  regressor=Sequential()
  regressor.add(Dense(units=6, kernel_initializer='uniform',activation='relu',input_dim=8))
  regressor.add(Dense(units=6, kernel_initializer='uniform',activation='relu'))
  regressor.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
  regressor.compile(optimizer='adam',loss='binary_crossentropy', metrics=['accuracy'])
  return regressor

regressor= KerasClassifier(build_fn=build_regressor,batch_size=10,epochs=100)
accuracies=cross_val_score(estimator=regressor,X=x_train,y=y_train,cv=10,n_jobs=-1)
mean=accuracies.mean()
variance=accuracies.std()

print(mean)
print(variance)

```
