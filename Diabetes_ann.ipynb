```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

dataset=pd.read_csv('diabetes.csv')
x=dataset.iloc[:,:-1]
y=dataset.iloc[:,-1]

#split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3)
#feature scaling
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)


import keras
from keras.models import Sequential
from keras.layers import Dense
predictor=Sequential()
predictor.add(Dense(units=7,kernel_initializer = 'uniform',activation='relu', input_dim=8))
predictor.add(Dense(units=7,kernel_initializer = 'uniform',activation='relu'))
predictor.add(Dense(units=7,kernel_initializer = 'uniform',activation='relu'))
predictor.add(Dense(units=1,activation='sigmoid'))
predictor.compile(optimizer='rmsprop',loss='binary_crossentropy',metrics=['accuracy'])
predictor.fit(x_train,y_train,batch_size=10,epochs=100)

y_pred=predictor.predict(x_test)
y_pred=(y_pred>0.5)


from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
from sklearn.metrics import accuracy_score
acc=accuracy_score(y_test,y_pred)

print(cm)
print(acc)
```
